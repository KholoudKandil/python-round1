### Step 0
Load the energy data from the file `Energy Indicators.xls`, which is a list of indicators of [energy supply and renewable electricity production](Energy%20Indicators.xls) from the [United Nations](http://unstats.un.org/unsd/environment/excel_file_tables/2013/Energy%20Indicators.xls) for the year 2013, and should be put into a DataFrame with the variable name of **energy**.

Keep in mind that this is an Excel file, and not a comma separated values file. Also, make sure to exclude the footer and header information from the datafile. The first two columns are unneccessary, so you should get rid of them, and you should change the column labels so that the columns are:

`['Country', 'Energy Supply', 'Energy Supply per Capita', '% Renewable']`

Convert `Energy Supply` to gigajoules (there are 1,000,000 gigajoules in a petajoule). For all countries which have missing data (e.g. data with "...") make sure this is reflected as `np.NaN` values.

Rename the following list of countries (for use in later questions):

```"Republic of Korea": "South Korea",
"United States of America": "United States",
"United Kingdom of Great Britain and Northern Ireland": "United Kingdom",
"China, Hong Kong Special Administrative Region": "Hong Kong"```

There are also several countries with numbers and/or parenthesis in their name. Be sure to remove these, 

e.g. 

`'Bolivia (Plurinational State of)'` should be `'Bolivia'`, 

`'Switzerland17'` should be `'Switzerland'`.

````python
# Import pandas
import pandas as pd
# to eliminate the false positive warning produced by step #3
# for further explaination see https://pandas.pydata.org/pandas-docs/stable/user_guide/indexing.html#returning-a-view-versus-a-copy
pd.options.mode.chained_assignment = None 


#Step 0

# Reading the energy indicator file and removing header  and footer and replacing "..." with NANs
cols=["Country", "Energy Supply", "Energy Supply per Capita", "%Renewable"]
energy = pd.read_excel('data/Energy Indicators.xls'
                      , header = 17
                      , nrows=227
                      , na_values=['...'])
                      
# droping first 2 cols
energy.drop(energy.columns[[0,1]], axis = 1, inplace = True)
#renaming columns
energy.columns=cols
#converting energy supply from petajoules to gigajoules
energy['Energy Supply'] = energy['Energy Supply'].apply(lambda x: x*1000000)


energy.head()
````

````python
#removing digits from country column
energy['Country'] = energy['Country'].str.replace('\d+','')
#removing paratenthis from country column
energy['Country'] = energy['Country'].str.replace(r"\(.*\)",'')

#Changing countries names
energy['Country'] = energy['Country'].replace({"Republic of Korea": "South Korea",
"United States of America": "United States",
"United Kingdom of Great Britain and Northern Ireland": "United Kingdom",
"China, Hong Kong Special Administrative Region": "Hong Kong"})


energy.head()
````

### Step 3
Join the three datasets: GDP, Energy, and ScimEn into a new dataset (using the intersection of country names). Use only the last 10 years (2006-2015) of GDP data and only the top 15 countries by Scimagojr 'Rank' (Rank 1 through 15). 

The index of this DataFrame should be the name of the country, and the columns should be ['Rank', 'Documents', 'Citable documents', 'Citations', 'Self-citations',
       'Citations per document', 'H index', 'Energy Supply',
       'Energy Supply per Capita', '% Renewable', '2006', '2007', '2008',
       '2009', '2010', '2011', '2012', '2013', '2014', '2015'].

*This step should yeild a DataFrame with 20 columns and 15 entries.*

````python
#Step 3

dfs = [energy,GDP,ScimEn]
all_data = pd.merge(pd.merge(energy,GDP,on='Country',how='inner'),ScimEn,on='Country',how='inner')
outer = pd.merge(pd.merge(energy,GDP,on='Country',how='inner'),ScimEn,on='Country',how='outer')

all_data = all_data[['Country','Rank', 'Documents', 'Citable documents', 'Citations', 'Self-citations', 'Citations per document', 'H index', 'Energy Supply', 'Energy Supply per Capita', '%Renewable', '2006', '2007', '2008', '2009', '2010', '2011', '2012', '2013', '2014', '2015']]
all_data.set_index('Country', inplace=True)
top_countries=all_data[all_data['Rank']<=15]
top_countries.sort_values(by='Rank',ascending=True, inplace=True)

top_countries
````

### Step 9
Create a new column with a 1 if the country's % Renewable value is at or above the median for all countries in the top 15, and a 0 if the country's % Renewable value is below the median.

*This step should return a series named `HighRenew` whose index is the country name sorted in ascending order of rank.*
